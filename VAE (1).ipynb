```python
import opendatasets as od

od.download('https://www.kaggle.com/datasets/alxmamaev/flowers-recognition', force =True)
```

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
import os
from tensorflow import keras
import pandas as pd
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
path = './flowers-recognition/flowers/'

images = []

for _, dir in enumerate(os.listdir(path)):
  for i, file in enumerate(os.listdir(os.path.join(path, dir))):
    image = load_img(os.path.join(path,dir,file))
    image = img_to_array(image)
    image = tf.image.resize(image, [102, 102])
    images.append(image)
    if i > 400:
      break
```

```python
images = np.asarray(images) / 255

x_train = images[:-10]
x_test = images[-10:]
```

```python
x_train.shape
```

```python
BATCH_SIZE = 64
HIDDEN_DIM = 256
LATENT_DIM = 32
IMAGE_SIZE = x_train.shape[1]*x_train.shape[2]*x_train.shape[3]

EPOCH = 10
LR = 0.0001
```

```python
from tensorflow.keras.layers import BatchNormalization, Dense, Conv2D, Flatten, Conv2DTranspose
```

```python
class VAE(keras.Model):
  def __init__(self, dim, **kwargs):
    super(VAE,self).__init__(**kwargs)
    h_dim = dim[0]
    z_dim = dim[1]

    self.fc1 = Dense(h_dim, activation='relu')
    self.fc2 = Dense(z_dim)
    self.fc3 = Dense(z_dim)
    self.fc4 = Dense(h_dim)
    self.fc5 = Dense(IMAGE_SIZE)
    self.encoder_norm = BatchNormalization()
    self.decoder_norm = BatchNormalization()

  def encode(self, x):

    x = self.encoder_norm(x)
    h = self.fc1(x)
    return self.fc2(h), self.fc3(h)

  def decode(self, z):
    h = self.fc4(z)
    h = self.fc5(h)
    return self.decoder_norm(h)

  def decode_sigmoid(self, z):
    return tf.nn.sigmoid(self.decode(z))

  def reparameterize(self, m, log):
    std = tf.exp(0.5 * log)
    eps = tf.random.normal(std.shape)
    return m + eps * std

  def call(self, input, training=None, mask=None):
    m, log = self.encode(input)
    z = self.reparameterize(m, log)
    x_log = self.decode(z)

    return x_log, m, log
```

```python
model = VAE([HIDDEN_DIM, LATENT_DIM])

model.build(input_shape=(BATCH_SIZE, IMAGE_SIZE))

print(model.summary())
```

```python
dataset = tf.data.Dataset.from_tensor_slices(x_train)
dataset = dataset.shuffle(BATCH_SIZE * 5).batch(BATCH_SIZE)
```

```python
loss_hist = []

opt = tf.keras.optimizers.Adam(learning_rate=LR)
batch_num = x_train.shape[0] // BATCH_SIZE

for epoch in range(EPOCH):
  for x in dataset:
    x = tf.reshape(x, [-1, IMAGE_SIZE])
    # print(x)
    with tf.GradientTape() as tape:
      x_reconstruct, m, log = model(x)
      reconstruction_loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=x, logits=x_reconstruct)
      reconstruction_loss = tf.reduce_sum(reconstruction_loss) / BATCH_SIZE
      kl_div = -0.5 * tf.reduce_sum(1. + log - tf.square(m) - tf.exp(log), axis=-1)
      kl_div = tf.reduce_mean(kl_div)
      loss = tf.reduce_mean(reconstruction_loss + kl_div)
    gr = tape.gradient(loss, model.trainable_variables)
    for g in gr:
      tf.clip_by_norm(g, 15)
    opt.apply_gradients(zip(gr, model.trainable_variables))
  loss_hist.append(float(reconstruction_loss))
  print(f"epoch : {epoch}/{EPOCH} | recon_loss : {float(reconstruction_loss):.2f} | KLDiv : {float(kl_div):.2f} | loss : {float(loss):.2f}")

```

```python
# loss_hist
plt.plot(loss_hist)
plt.show()
```

```python
def out(z):
  out = model.decode_sigmoid(z)
  out = tf.reshape(out , (-1, x_train.shape[1],x_train.shape[2],x_train.shape[3]))
  return out
```

```python
m, log  = model.encode(x_test.reshape(-1, IMAGE_SIZE))
z = model.reparameterize(m, log)
o = out(z)
num = 10
plt.figure(figsize=(20,8))
for idx in range(num):
  ax = plt.subplot(2, num, idx + 1)
  plt.imshow(x_test[idx].reshape(x_test.shape[1:]), cmap='gray')
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)

for idx in range(num):
  ax = plt.subplot(2, num, num + idx + 1)
  plt.imshow(o[idx])
  ax.get_xaxis().set_visible(False)
  ax.get_yaxis().set_visible(False)

plt.show()
```
